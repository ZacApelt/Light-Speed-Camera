# Determining the transfer function of the DSOX2004a's analog front end

Using the laser's optical trigger pulse because this signal is well characterised and has a rise time of < 800ps

```python
import numpy as np
import matplotlib.pyplot as plt

# x is the signal
x_0 = -3.694e-2
x_max = 4.347

# find the rise time between 10% and 90% of the max signal
x_10 = x_0 + 0.1 * (x_max - x_0)
x_90 = x_0 + 0.9 * (x_max - x_0)
print(x_10, x_90)

t_10 = -1.934e-9
t_90 = 1.581e-9

T_r = t_90 - t_10
print("Rise time:", T_r)

# signal actually has a rise time of 800ps
t_rise_scope = np.sqrt(T_r**2 - (800e-12)**2)

print("rise time of scope:", t_rise_scope)

# for a 4th order butterworth filter:

f_c = 0.387 / t_rise_scope

print("Bandwidth:", f_c)
```

The analog front end can be modeled as a 4th order butterworthe filter. This has a transfer function:

H(s) = \frac{1}{(s^2+0.7654s+1)(s^2+1.8478s+1)}
